```python
import requests
import os
from PIL import Image
```

```python
wd = './val/'
names = os.listdir(wd)
```

```python
for filename in names:
    filename = wd + filename
    img = Image.open(filename)
    display(img)

    with open(filename, 'rb') as f:
        r = requests.post('http://localhost:5000/', files={'file': f})
    print(r.text)  
```
